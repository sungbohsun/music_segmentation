```python
def match (rule,labels):
    import re
    x = []
    for label in labels:
        mach = re.findall(".{}..*".format(rule), label)
        if mach != []:
            x.extend(mach)
    return list(set(x))

# 1、引子 （intro）
# 2、主歌 （verse）
# 3、連接段 （Bridge）
# 4、結語 （outro）
# 5、休息 (break)
# 6、副歌  (Refrain)
# 7、沉默  (silece)
typs = ['intro','verse','bridge','outro','break','Refrain','silece']
serchs = ['ntr','ers','ridg','utr','rea','efrai','ile']
results = [match(serch,labels) for serch in serchs]

from glob import glob
labels = []
count = []

w = open('annotations/Labels.txt','w')

for i in range(1,13):
    album = i
    files = glob('DL/the_beatles_label/{0:02d}*/*'.format(album))
    
    for file in files:    
        song = int(file.split('/')[3][:2])
        t = open('annotations/Labels/{0:02d}_{1:02d}.txt'.format(album,song),'w')
        w.write('Labels/{0:02d}_{1:02d}.txt\n'.format(album,song))
        with open(file) as f:
            content = f.readlines()
        contents = [x.strip() for x in content]

        for content in contents :
            tmp = content.split(' ')[2]
            if tmp in sum(results,[]):
                for typ,result in zip(typs,results):
                    if tmp in result:
                        #print(tmp,'->',typ)
                        tmp = typ
            else:
                tmp = 'other'
            t.write(content.split(' ')[0]+' '+content.split(' ')[1]+' '+tmp+'\n')
        t.close()
        f.close()
w.close()
```

```python
# make ground truth in msaf
import json
from glob import glob
files = glob('annotations/Labels/*.txt')
for file in files:
    with open('references/01-Sargon-Mindless.jams_copy') as f:
        data = json.load(f)

    t = open(file,'r')
    result = []
    txts = t.read()
    tmp = txts.split('\n')
    for i in range(len(tmp)):
        dic = {}
        if tmp[i] != '':
            dic['time'] = float(tmp[i].split(' ')[0])
            dic['duration'] = float(tmp[i].split(' ')[1])-float(tmp[i].split(' ')[0])
            dic['confidence'] = None
            dic['value'] = tmp[i].split(' ')[2]
            result.append(dic)

    data['annotations'][0]['data'] = result

    with open('references/{}.jams'.format(file.split('/')[2][:5]), 'w') as json_file:
        json.dump(data, json_file)
```

```python
from __future__ import print_function
from glob import glob
import msaf
import json
import librosa
import seaborn as sns
import IPython.display
from ipywidgets import interact

sns.set(style="dark")
audio_files = sorted(glob('audio/*'))

# # First, let's list all the available boundary algorithms
# print('boundary:',msaf.get_all_boundary_algorithms())
# print('label:',msaf.get_all_label_algorithms())
# print('features:',*msaf.features_registry)
def score (audio_file,est):
    with open('references/{}.jams'.format(audio_file.split('/')[1][:5])) as f:
        data = json.load(f)

    ref = [L['time'] for L in data['annotations'][0]['data']]
    sc = sum([min([abs(r - e) for e in est])for r in ref])+(len(est)-len(ref))*len(ref)
    return sc
# Obtain the default configuration

def _run_msaf(audio_file,plot=True):
    boundaries, labels = msaf.process(audio_file, 
                                      feature='pcp',
                                      boundaries_id="foote", 
                                      labels_id="fmc2d", plot=plot)
    sc = score(audio_file,boundaries)
    return sc

def _run_msaf＿adj(audio_file,feature,M_gaussian,plot=True):
    config = msaf.io.get_configuration(feature = feature, 
                                       annot_beats=False, 
                                       framesync=False, 
                                       boundaries_id='sf', 
                                       labels_id='fmc2d')
    config["M_gaussian"] = M_gaussian
    boundaries,labels = msaf.process(audio_file, 
                           n_jobs = 36,
                           feature=feature,
                           config=config, 
                           plot=plot)
    sc = score(audio_file,boundaries)
    return sc
```

```python
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings("ignore")
sc_all = []
feature = ['cqt','mfcc','pcp','tonnetz','tempogram']
for f in tqdm(feature):
    sc = []
    for audio_file in tqdm(audio_files):
        sc_adj = []
        for M in range(0,61,5):
            tmp = _run_msaf_adj(audio_file,f,M_gaussian = M,plot = False)
            sc_adj.append(tmp)
        sc.append(sc_adj)
    sc_all.append(sc)
```
